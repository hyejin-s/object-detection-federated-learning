```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("train2017.pickle", "rb") as f:
    train_dic = pickle.load(f)
with open("low_map_365.pickle", "rb") as f:
    low_dic = pickle.load(f)

ratio_list = np.zeros(80)

for k, v in train_dic.items():
    ratio_list[k] = low_dic[k] / v


print(*zip(np.arange(80),ratio_list))
```

```python
print(train_dic[8], low_dic[8])
```

```python
plt.figure(figsize=(15, 10))
bars = plt.bar(np.arange(80), ratio_list)
bars[56].set_color('r')
bars[60].set_color('r')
plt.xlabel("Class", fontsize=20)
plt.ylabel("The error ratio of classes", fontsize=20)
plt.title("Error images / Training images", fontsize=25)
plt.show()
```

```python
f = open('./val.txt', 'r')
lines = [line.strip() for line in f.readlines()]
f.close()
```

```python
lines[0].split()
class_list = np.roll([l.split()[0] for l in lines], -1)
mAP50_list =  np.roll([float(l.split()[-2]) for l in lines],-1)
mAP95_list = np.roll([float(l.split()[-1]) for l in lines],-1)
```

```python
plt.figure(figsize=(15, 10))
bars = plt.bar(np.arange(81), mAP95_list)
bars[80].set_color('b')
bars[56].set_color('r')
bars[60].set_color('r')
plt.xlabel("Class", fontsize=20)
plt.ylabel("mAP50:95", fontsize=20)
plt.title("Validation results", fontsize=25)
plt.show()
```

```python
f = open('./yolov5s_val.txt', 'r')
y_lines = [line.strip() for line in f.readlines()]
f.close()
y_class_list = np.roll([l.split()[0] for l in y_lines], -1)
y_mAP50_list =  np.roll([float(l.split()[-2]) for l in y_lines],-1)
y_mAP95_list = np.roll([float(l.split()[-1]) for l in y_lines],-1)
plt.figure(figsize=(15, 10))
bars = plt.bar(np.arange(81), y_mAP95_list)
bars[80].set_color('b')
bars[56].set_color('r')
bars[60].set_color('r')
bars[8].set_color('y')
plt.xlabel("Class", fontsize=20)
plt.ylabel("mAP50:95", fontsize=20)
plt.title("Yolov5s validation results", fontsize=25)
plt.show()
```

```python
mAP_ratio_list = np.divide(mAP95_list, y_mAP95_list)
plt.figure(figsize=(15, 10))
bars = plt.bar(np.arange(81), mAP_ratio_list)
bars[80].set_color('b')
bars[56].set_color('r')
bars[60].set_color('r')
plt.axhline(y = 0.9, color='r')
plt.xlabel("class", fontsize=20)
plt.ylabel("ratio", fontsize=20)
plt.title("mAP ratio (pretrained, yolov5s)", fontsize=25)
plt.show()
```
